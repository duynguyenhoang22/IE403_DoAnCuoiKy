```python
import pandas as pd
```

```python
df1 = pd.read_csv('../GgSheetData_preprocessed.csv')
df2 = pd.read_csv('../labeled_sms_data.csv')

df = pd.concat([df1, df2])
```

```python
df['label'] = df['label'].astype(int)
```

```python
df.to_csv('../dataset.csv', index=False)
```

```python
df_spam = df[df['label'] == 1]
df_ham = df[df['label'] == 0]
```

```python
df_spam.to_csv('spam.csv', index=False)
df_ham.to_csv('ham.csv', index=False)
```
